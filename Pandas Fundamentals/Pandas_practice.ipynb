## Merging & Concat in python pandas

```python
import pandas as pd 
var_m=pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var_n=pd.DataFrame({"A":[1,2,3,4,5],"C":[21,22,23,24,25]})
pd.merge(var_m,var_n,on="A")# this is common part in both of them
```

```python
pd.merge(var_m,var_n,how="inner")
```

```python
pd.merge(var_m,var_n,how="outer")
```

```python
pd.merge(var_m,var_n,how="left")
```

```python
pd.merge(var_m,var_n,how="right")
```

```python
pd.merge(var_m,var_n,how="outer",indicator=True)
```

```python
pd.merge(var_m,var_n,left_index=True,right_index=True)
```

## Concat

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,12,13,14])
pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2],axis=1)#row wise 
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([d1,d2],axis=0)# colonwise 
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[3,5],"B":[23,24]})
pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[3,5],"B":[23,24]})
pd.concat([d1,d2],axis=1,join="inner")
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[3,5],"B":[23,24]})
pd.concat([d1,d2],axis=1,join="outer")
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"A":[3,5],"B":[23,24]})
pd.concat([d1,d2],axis=1,keys=["p1","p2"])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
d2=pd.DataFrame({"C":[3,5],"D":[23,24]})
pd.concat([d1,d2],axis=1)
```
